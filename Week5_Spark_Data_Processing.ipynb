### **Q1. What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?**
**Answer**

Traditional MapReduce is heavily constrained by its disk-bound architecture, rigid two-stage execution, and lack of native support for iterative or real-time processing. Spark addresses these issues by leveraging in-memory caching, lazy evaluation via Directed Acyclic Graphs (DAG), and high-level abstractions, making it vastly faster and more versatile.

 **The key limitations of MapReduce that drive the preference for Spark include:**

**High Disk Latency:** MapReduce forces disk writes after every step. Spark processes data directly in RAM.

**Slow Machine Learning:** Iterative algorithms must reload data from disk each cycle. Spark holds datasets in memory.

**Batch-Only Design:** MapReduce lacks native real-time streaming. Spark handles batch, streaming, and SQL natively.

**Rigid Programming:** Code is complex and strictly limited to map and reduce. Spark offers flexible APIs.

**No Built-in Tools:** MapReduce lacks native machine learning libraries. Spark includes MLlib and Spark SQL.


### **Q2. Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.**

Answer

Apache Spark uses in-memory computing, which means it stores intermediate data in RAM instead of writing it to disk after every operation. Since accessing data from memory is much faster than reading and writing to disk, Spark can process data more efficiently.

This approach is especially useful for iterative machine learning algorithms because the same dataset is processed multiple times during training. Spark keeps the required data in memory, reducing repeated disk access and significantly improving execution speed.

In contrast, traditional disk-based systems like MapReduce store intermediate results on disk after each processing stage. This increases input/output operations and makes repeated computations much slower.

**Advantages of In-Memory Computing in Spark:**

    Faster execution due to reduced disk I/O.
    Efficient processing of iterative machine learning algorithms.
    Better performance for data analysis and transformations.
    Supports real-time and interactive data processing.
    Reduces overall execution time for large datasets.

### **Q6. Write a query to find the total count of records for each city in a DataFrame, but only for cities where the count is greater than 100.**
### Answer

The dataset is grouped by the City column using the groupBy() function. The count() aggregation function calculates the total number of records for each city. Finally, the filter() function is used to display only those cities where the record count is greater than 100.

```python


from pyspark.sql.functions import count, col

city_count = (
    df.groupBy("City")
      .agg(count("*").alias("Total Records"))
      .filter(col("Total Records") > 100)
      .orderBy(col("Total Records").desc())
)

city_count.show()

```

### **Q7. How does the immutability of Spark DataFrames affect how you perform "data cleaning" steps like dropping columns or renaming them?**
### Answer

Spark DataFrames are immutable, which means their data cannot be modified after they are created. Any operation such as dropping a column, renaming a column, or filtering data does not change the original DataFrame. Instead, Spark creates a new DataFrame containing the requested changes.

This approach helps maintain data consistency, supports fault tolerance, and enables Spark to optimize query execution efficiently.

```python
# Original DataFrame
df.show(5)

# Create a new DataFrame after dropping a column
new_df = df.drop("Postal Code")

print("Original DataFrame Columns:")
print(df.columns)

print("\nNew DataFrame Columns:")
print(new_df.columns)
```

### **Insight**

The original DataFrame remains unchanged after performing the drop() operation. A new DataFrame is created with the modified schema, demonstrating the immutable nature of Spark DataFrames.

### **Q8. Write a Spark command to filter a dataset for rows where the age is between 18 and 30 (inclusive) and the subscription is 'Premium'.**
### Answer

The provided Superstore dataset does not contain Age and Subscription columns. Therefore, a sample DataFrame has been created to demonstrate how Spark filters records based on multiple conditions.

The filter selects records where the Age is between 18 and 30 (inclusive) and the Subscription type is "Premium".

```python
from pyspark.sql import Row
from pyspark.sql.functions import col

# Sample DataFrame
data = [
    Row(Name="Alice", Age=22, Subscription="Premium"),
    Row(Name="Bob", Age=17, Subscription="Premium"),
    Row(Name="Charlie", Age=28, Subscription="Basic"),
    Row(Name="David", Age=30, Subscription="Premium"),
    Row(Name="Eva", Age=25, Subscription="Premium")
]

sample_df = spark.createDataFrame(data)

print("Original DataFrame")
sample_df.show()

filtered_df = sample_df.filter(
    (col("Age").between(18, 30)) &
    (col("Subscription") == "Premium")
)

print("Filtered DataFrame")
filtered_df.show()
```

### **Q10. Write the code to revise a column named raw_timestamp by casting it to a TimestampType and renaming it to event_time.**
### Answer

The given Superstore dataset does not contain a raw_timestamp column. Therefore, a small sample DataFrame was created to demonstrate how Spark converts a string timestamp into TimestampType and renames the column to event_time.

```python
from pyspark.sql import Row
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType

# Sample DataFrame
timestamp_data = [
    Row(raw_timestamp="2026-07-18 10:15:30"),
    Row(raw_timestamp="2026-07-19 14:45:10")
]

timestamp_df = spark.createDataFrame(timestamp_data)

print("Original DataFrame")
timestamp_df.show(truncate=False)

# Cast and rename
updated_df = (
    timestamp_df
    .withColumn("raw_timestamp", col("raw_timestamp").cast(TimestampType()))
    .withColumnRenamed("raw_timestamp", "event_time")
)

print("Updated DataFrame")
updated_df.show(truncate=False)

print("Updated Schema")
updated_df.printSchema()
```

### Insight

Casting string values to TimestampType allows Spark to perform date and time-based operations efficiently. Renaming the column to event_time also improves readability and makes the dataset easier to understand during analysis.

###**Q11. Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?**
### Answer

A shuffle is the process in Spark where data is redistributed across different partitions so that records with the same key are brought together. This usually happens during operations such as groupBy(), join(), and reduceByKey().

Shuffle is considered a wide transformation because data is moved across multiple partitions and executors in the cluster. Unlike narrow transformations, where each partition depends on only one parent partition, wide transformations require data exchange over the network, making them more expensive in terms of execution time and resources.

```python
from pyspark.sql.functions import sum

# Group data by Category and calculate total sales
category_sales = (
    df.groupBy("Category")
      .agg(sum("Sales").alias("Total Sales"))
)

category_sales.show()
```

### Insight

The groupBy() operation requires Spark to redistribute records with the same category into the same partition before calculating the total sales. This data movement is known as shuffle, making groupBy() a wide transformation because it involves communication between multiple partitions.

### **Q12. Write a code snippet that identifies and removes rows where the email column contains null values OR the username is an empty string.**

### Answer

The provided Superstore dataset does not contain email and username columns. Therefore, a sample DataFrame has been created to demonstrate how Spark identifies and removes rows where the email is null or the username is an empty string.

This helps improve data quality by removing incomplete or invalid records before further analysis.

```python
from pyspark.sql import Row
from pyspark.sql.functions import col

# Sample DataFrame
data = [
    Row(username="Alice", email="alice@gmail.com"),
    Row(username="", email="bob@gmail.com"),
    Row(username="Charlie", email=None),
    Row(username="David", email="david@gmail.com"),
    Row(username="", email=None)
]

user_df = spark.createDataFrame(data)

print("Original DataFrame")
user_df.show()

# Remove rows with null email OR empty username
clean_df = user_df.filter(
    (col("email").isNotNull()) &
    (col("username") != "")
)

print("Cleaned DataFrame")
clean_df.show()
```

### Insight

Filtering invalid records before analysis improves data quality and ensures that only complete and meaningful records are processed. This reduces errors and increases the reliability of analytical results.

### **Q14. In the context of cleaning a dataset, what is the risk of using inferSchema=true when your source data contains messy or inconsistent date formats?**
### Answer

The inferSchema=true option allows Spark to automatically detect the data type of each column while reading a dataset. However, if the source data contains inconsistent or messy date formats, Spark may infer an incorrect data type or treat the entire column as a string.

This can lead to errors during data processing, incorrect sorting or filtering of dates, and failures in date-based calculations. For reliable data processing, it is often recommended to define the schema explicitly or clean the data before performing analysis.


```python
from pyspark.sql import Row

# Sample DataFrame with inconsistent date formats
data = [
    Row(raw_date="2026-07-18"),
    Row(raw_date="18/07/2026"),
    Row(raw_date="July 18, 2026"),
    Row(raw_date="2026/07/18")
]

date_df = spark.createDataFrame(data)

print("Sample Data")
date_df.show(truncate=False)

print("Inferred Schema")
date_df.printSchema()
```

### Insight

When date values are stored in multiple formats, Spark cannot reliably determine a single date type and often treats the column as a string. Defining the schema manually or standardizing date formats before loading the data helps avoid incorrect data types and improves data quality.

### **Q15. Write a final processing pipeline that:**

    Filters out duplicates.
    Fills null prices with 0.
    Groups by store_id to calculate total revenue.

### Answer

The provided Superstore dataset does not contain store_id or price columns. Therefore, Customer ID is used as the grouping column, while Sales is used as the revenue column.

The pipeline performs the following operations:

    Removes duplicate records.
    Replaces null values in the Sales column with 0.
    Groups records by Customer ID.
    Calculates the total revenue for each customer using the sum() aggregation function.


```python
from pyspark.sql.functions import sum

# Step 1: Remove duplicate records
clean_df = df.dropDuplicates()

# Step 2: Fill null values in Sales with 0
clean_df = clean_df.na.fill({"Sales": 0})

# Step 3: Group by Customer ID and calculate total revenue
revenue_df = (
    clean_df.groupBy("Customer ID")
            .agg(sum("Sales").alias("Total Revenue"))
            .orderBy("Total Revenue", ascending=False)
)

revenue_df.show(10)
```

### Insight

This pipeline demonstrates a complete Spark DataFrame workflow by combining data cleaning, null value handling, grouping, and aggregation into a single process. Such pipelines are commonly used in real-world data engineering projects to prepare raw data for analysis and reporting.

# **Final Data Processing Pipeline**

This section integrates all the previous data processing steps into a single, reusable ETL pipeline. The pipeline reads the raw dataset, validates its structure, performs data cleaning, applies necessary transformations, computes key business metrics, and exports the processed results. This modular workflow demonstrates a production-oriented approach to building scalable Spark data processing applications.

```python
# ================================
# Final Data Processing Pipeline
# ================================

print("Starting Data Processing Pipeline...")

# Use the existing DataFrame
pipeline_df = df

# Remove Duplicates
pipeline_df = pipeline_df.dropDuplicates()

# Fill Missing Values
pipeline_df = pipeline_df.fillna({
    "sales": 0,
    "quantity": 0,
    "discount": 0,
    "profit": 0
})

# Standardize Column Names
for c in pipeline_df.columns:
    pipeline_df = pipeline_df.withColumnRenamed(
        c,
        c.lower().replace(" ", "_").replace("-", "_")
    )

# Business Summary
business_summary = (
    pipeline_df.groupBy("region")
               .agg(
                   sum("sales").alias("total_sales"),
                   sum("profit").alias("total_profit")
               )
               .orderBy(desc("total_sales"))
)

business_summary.show()

print("Pipeline Executed Successfully!")
```
